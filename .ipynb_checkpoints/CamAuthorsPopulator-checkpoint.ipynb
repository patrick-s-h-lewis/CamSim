#CamAuthors Populator
Takes data from camharvest collection in camsim database and generates the cam authors database

```python
import pymongo
from pymongo import MongoClient
mongo_url = 'mongodb://localhost:27017/'
db = 'CamSim'
coll_in = 'CamHarvest'
coll_out = 'CamAuthors'
client = MongoClient(mongo_url)
ch = client[db][coll_in]
ca = client[db][coll_out]
```

```python
cursor = ch.find()
ind = 0
for doc in cursor:
    authors = doc['authors']
    for author in authors:
        ca.update_one({'name':author},
                     {'$push':{'pubs':doc}
                
            },
                     upsert=True)
    if (ind%100 == 0):
        print('visited '+str(ind)+' records')
    ind+=1
    
```
