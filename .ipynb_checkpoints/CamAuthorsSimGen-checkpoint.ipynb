#CamAuthorsSimGen
generate similarities between cam authors

```python
import gensim
import pymongo
from pymongo import MongoClient

mongo_url = 'mongodb://localhost:27017/'
db = 'CamSim'
coll = 'CamAuthors'
client = MongoClient(mongo_url)
ca = client[db][coll]

sim_name = 'SimEngine'
model = gensim.models.Word2Vec.load(sim_name)
```

```python
import time
cur1=ca.find()
for rec1 in cur1[0:1]:
    corp1 = rec['corpus']
    sims = []
    x = time.time()
    cur2 = ca.find()
    for rec2 in cur2:
        corp2 = rec2['corpus']
        s = model.n_similarity(corp1.split(),corp2.split())
        sims.append({rec2['name']:s})
    ca.update_one({'name':rec1['name']},{'$set':{'sims':sims}})
    print (unicode(str(time.time() - x)) + u' seconds to produce sims for '+rec1['name'])
```
